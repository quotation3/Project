```python
import pandas as pd
from ekonlpy.sentiment import MPCK
```

```python
data=pd.read_csv("./minutes/minutes_preprocessed.csv")
data
```

```python
data_sentence=data.copy()[["date","Financial Markets"]]
data_sentence["sentence_ls"]=data_sentence["Financial Markets"].str.split("@@@")
data_sentence
```

```python
drop_data=data_sentence[["date","sentence_ls"]].dropna()
```

```python
splited_data=pd.DataFrame(pd.DataFrame(drop_data["sentence_ls"].tolist()).stack())
splited_data["date"]=[drop_data.iloc[x[0]]["date"] for x in splited_data.index]
splited_data.columns=["sentences","date"]
splited_data
```

```python
def text2ngram(text) :
    mpck = MPCK()
    tokens = mpck.tokenize(text)
    ngrams = mpck.ngramize(tokens)
                
    return tokens+ ngrams
```

```python
splited_data["ngram"]=splited_data["sentences"].apply(text2ngram)
```

```python
splited_data.to_csv("./minutes/sentences_ngram_sec3.csv")
```
