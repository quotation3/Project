```
import pandas as pd
```

```
yellow = pd.read_csv('/content/drive/My Drive/yello+accident.csv의 사본', encoding='euc-kr')
```

```
yellow.head()
```

```
yellow['before'] = ''
yellow['after']=''
for i in range(len(yellow)):
    a = yellow['전'][i]/(yellow['year'][i]-2009)
    yellow['before'][i] = yellow['before'][i].replace('', str(a))
    b = yellow['후'][i]/(2019-yellow['year'][i])
    yellow['after'][i] = yellow['after'][i].replace('', str(b))
```

```
import matplotlib.pyplot as plt
```

```
x = yellow['before']
y = yellow['after']
plt.scatter(x,y)
```

```
df = yellow[['before','after']]
```
