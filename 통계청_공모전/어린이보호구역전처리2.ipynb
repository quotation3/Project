```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Inyong Kim/Desktop/통계청공모전/전처리데이터/어린이보호구역.csv', encoding='euc-kr')
df
```

```python
for i in range(len(df['보호구역도로폭'])):
    if df['보호구역도로폭'][i] <= 11:
        df['차수'][i] = 1
    elif df['보호구역도로폭'][i] <= 16:
        df['차수'][i] = 2
    elif df['보호구역도로폭'][i] > 16:
        df['차수'][i] = 3
```

```python
for i in range(len(df['차수'])):
    if df['차수'][i]==1:
        df['제한속도'][i]=30
```

```python
df.to_csv('C:/Users/Inyong Kim/Desktop/통계청공모전/전처리데이터/어린이보호구역_제한속도.csv', index=False, encoding='euc-kr')
```
