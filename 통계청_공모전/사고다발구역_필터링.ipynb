```python
import glob

post_path = "C:\\Users\\Inyong Kim\\Desktop\\사고정보_전처리\\서울\\*"
post_file_list = glob.glob(post_path)
post_file_list
```

```python
import pandas as pd
```

```python
df = pd.read_csv(post_file_list[0], encoding='euc-kr')
for i in range(1,len(post_file_list)):
    df2 = pd.read_csv(post_file_list[i], encoding='euc-kr')
    df = pd.concat([df2,df],axis=0)
```

```python
df.reset_index()
```

```python
df_sort = df[['사고장소','사고일시','시군구']]
df_sort
```

```python
df_sort['사고장소'].value_counts()
```

```python
cnt_dict = dict(df_sort['사고장소'].value_counts())
cnt_dict
```

```python
keys = []
values = []

for key in cnt_dict.keys():
    keys.append(key)
for value in cnt_dict.values():
    values.append(value)
```

```python
spot = pd.DataFrame(keys)
count = pd.DataFrame(values)
table = pd.concat([spot,count], axis=1)
table.columns = ['대상시설명', 'count']
table
```

```python
# 사고 3회이상 구역 필터링
final = table['count'] > 2
cnt_table = table[final]
cnt_table
```

```python
info = pd.read_csv('C:\\Users\\Inyong Kim\\Desktop\\어린이보호구역_차선_제한속도+카+옐.csv', encoding='euc-kr')
for i in range(len(info['대상시설명'])):
    info['대상시설명'][i] = info['대상시설명'][i].replace('초등학교','초').replace('초교','초').replace('초등','초')
```

```python
# 기존 어린이보호구역 정보와 merge
zone = pd.merge(info,cnt_table, on='대상시설명', how='inner')
```

```python
camera = zone['카메라설치여부'] == 1
no_camera = zone['카메라설치여부'] == 0
yellow = zone['옐로카펫설치여부'] == 1
no_yellow = zone['옐로카펫설치여부'] == 0
```

```python
# 카메라, 옐로카펫 둘다 있는 구역
both = zone[camera&yellow]
both
```

```python
# 카메라만 있는 구역
camera_only = zone[camera & no_yellow]
camera_only
```

```python
# 옐로카펫만 있는 구역
yellow_only = zone[no_camera & yellow]
yellow_only
```

```python
# 카메라, 옐로카펫 둘 다 없는 구역
nothing = zone[no_camera & no_yellow]
nothing
```

```python
both.to_csv('C:\\Users\\Inyong Kim\\Desktop\\both.csv', index=False, encoding='utf-8')
camera.to_csv('C:\\Users\\Inyong Kim\\Desktop\\caemra.csv', index=False, encoding='utf-8')
yellow.to_csv('C:\\Users\\Inyong Kim\\Desktop\\yellow.csv', index=False, encoding='utf-8')
nothing.to_csv('C:\\Users\\Inyong Kim\\Desktop\\nothing.csv', index=False, encoding='utf-8')
```
