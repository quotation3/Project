```python
width = []
for i in seoul['보호구역도로폭']:
    if type(i) == float:
        width.append(i)
    elif type(i) == int:
        width.append(i)
    elif type(i) == str:
        width.append(i.split("~")[1])
    else:
        width.append(i)
```

```python
wd = pd.DataFrame(width, columns = ['도로폭'])
```

```python
seoul = pd.concat([seoul,wd],axis=1, ignore_index=False)
```
