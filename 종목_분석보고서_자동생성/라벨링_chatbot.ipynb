```python
import pandas as pd
import json
import numpy as np
```

```python
news = pd.read_csv('C:\\Users\\Inyong Kim\\Desktop\\Sentence_Word.csv', encoding='utf-8')
news
```

```python
stock = pd.read_csv('C:\\Users\\Inyong Kim\\Desktop\\samsung_stock_price.csv',encoding = 'utf-8')
```

```python
news.rename(columns = {'Date': 'date'}, inplace = True)
```

```python
stock.rename(columns = {'Date': 'date'}, inplace = True)
```

```python
from datetime import datetime, timedelta

start_date = datetime.strptime('20010101', '%Y%m%d')
end_date = datetime.strptime('20201005', '%Y%m%d')

str_date_list = []
while start_date.strftime('%Y%m%d') != end_date.strftime('%Y%m%d'): # start_date가 end_date랑 같아지기 전까지 1씩 늘어나면서 리스트에 추가
    str_date_list.append(start_date.strftime('%Y.%m.%d'))
    start_date += timedelta(days=1)
str_date_list.append(end_date.strftime('%Y.%m.%d')) # end_date 포함안됐으므로 따로 추가

date_list=np.array([str_date_list])
all_date  = pd.DataFrame(date_list.T, columns=['date'])

all_date['date'] = pd.to_datetime(all_date['date'], format="%Y.%m.%d")
stock['date'] = pd.to_datetime(stock['date'], format="%Y.%m.%d")

df = pd.merge(all_date,stock, on='date', how='outer') # 기존주가데이터와 전체날짜를 date기준으로 outer 조인
df
```

```python
new_df = df[['date','Close','Volume']]
new_df
```

```python
new_df = new_df.set_index('date')
new_df['Close'] = new_df['Close'].interpolate(method='time')
new_df['Volume'] = new_df['Volume'].interpolate(method='time')
new_df
```

```python
new_df.reset_index(drop=False, inplace=True)
new_df
```

```python
new_df['stock_ma']=''
```

```python
# 7일간의 이동평균
for i in range(7, len(new_df)):
    new_df['stock_ma'][i]=new_df['Close'][i-7:i+1].mean()
new_df
```

```python
new_stock = new_df[7:]
new_stock['label']=''
new_stock.reset_index(drop=True, inplace=True)
new_stock
```

```python
for i in range(len(new_stock)):
    if new_stock['Close'][i] - new_stock['stock_ma'][i] > 0:
        new_stock['label'][i] = new_stock['label'][i].replace('','2')
    elif new_stock['Close'][i] - new_stock['stock_ma'][i] < 0:
        new_stock['label'][i] = new_stock['label'][i].replace('','0')
    elif new_stock['Close'][i] - new_stock['stock_ma'][i] == 0:
        new_stock['label'][i] = new_stock['label'][i].replace('','1')
```

```python
new_stock.label = new_stock.label.shift(-7)
new_stock
```

```python
stock_label = new_stock[['date','label']]
```

```python
stock_label['date'] = pd.to_datetime(stock_label['date'], format="%Y-%m-%d")
news['date'] = pd.to_datetime(news['date'], format="%Y-%m-%d")
```

```python
label_for_chatbot = pd.merge(news, stock_label, on='date', how='inner')
label_for_chatbot
```

```python
label_for_chatbot.to_csv('C:\\Users\\Inyong Kim\\Desktop\\label_for_chatbot.csv', index=False, encoding='utf-8')
```
