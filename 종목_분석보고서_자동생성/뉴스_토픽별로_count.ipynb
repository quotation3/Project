```python
import pandas as pd
import json
import numpy as np
```

```python
news = pd.read_json('C:/Users/user/Desktop/news_df.json')
news
```

```python
topic = pd.read_json('C:/Users/user/Desktop/3_topic.json')
topic
```

```python
lis = []

for a,b,c in topic['topic']:
    tmp = []
    for text in news['text']:
        if a in text and b in text and c in text:
            tmp.append(1)
        else:
            tmp.append(0)
            
    lis.append(tmp)
lis
```

```python
topic_count = pd.DataFrame(np.array(lis).T, columns=['topic'+str(n+1) for n in range(len(topic))])
```

```python
total = pd.concat([news,topic_count], axis=1)
total
```

```python
by_date = []

for i in range(3):
    by_date.append(total['topic{}'.format(i+1)].groupby(total['date']).sum())
```

```python
df = pd.DataFrame(by_date)
df
```

```python
count_df = df.transpose()
count_df
```
