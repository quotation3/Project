# 뉴스 nltk로 전처리 후 n-gram화 하기

```python
import pandas as pd
```

```python
brazilnews = pd.read_csv('C:/Users/user/Desktop/globo/brazilnews_final.csv')
brazilnews.sort_values(by=['date'], axis=0, inplace=True)
brazilnews['total'] = brazilnews['main_title']+brazilnews['sub_title']+brazilnews['content']
brazilnews.dropna(axis=0, inplace=True)
brazilnews.reset_index(drop=True, inplace=True)
```

```python
brazilnews['total']=brazilnews['total'].replace(' “','').replace('” ','')
brazilnews['total']=brazilnews['total'].str.strip()
```

```python
import nltk
nltk.download('punkt')
import nltk.data
sent_tokenizer=nltk.data.load('tokenizers/punkt/portuguese.pickle')
# from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
nltk.download('stopwords')
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("portuguese")
```

### stopwords 편집

```python
stopwords = nltk.corpus.stopwords.words('portuguese')
stopwords.append('"')
stopwords.append('(')
stopwords.append(')')
stopwords.append('“')
stopwords.append('”')
stopwords.append('``')
stopwords.append("''")
stopwords.append(":")
stopwords.append("[")
stopwords.append("]")
stopwords.append(",")
stopwords.append("$")
stopwords.append("1")
stopwords.append("10")
stopwords.append("100")
stopwords.append("0")
stopwords.append("%")
stopwords.append("2")
stopwords.append("5")
stopwords.append("12")
stopwords.append("2017")
stopwords.append("2018")
stopwords.append("2019")
stopwords.append("2020")
stopwords.append("&")
stopwords.append("3")
stopwords.append("!")
stopwords.append("?")
stopwords.append("-")
stopwords.append("--")
stopwords.append("...")
stopwords.append("s")
stopwords.append(".")
stopwords
```

### 전처리 및 n-gram 만들기

word_tokenize->stopwords->stemming->n-gram 순

```python
gram1 = []

for i in brazilnews['total']:
    tmp = []
    word_tokens = word_tokenize(i)
    result = []
    for token in word_tokens:
        if token not in stopwords:
            result.append(token)
    stem_lis = []
    for word in result:
        stem_lis.append(stemmer.stem(word))
    n_grams = ngrams(stem_lis,1)
    results = [';'.join(grams) for grams in n_grams]
    gram1.append(results)
```

### 데이터프레임화 한 후 브라질 뉴스와 결합

```python
a = pd.DataFrame(zip(gram1), columns=['one'])
one = []
for i in a['one']:
    one.append(','.join(i))
```

```python
b = pd.DataFrame(zip(gram2), columns=['two'])
two = []
for i in b['two']:
    two.append(','.join(i))
```

```python
c = pd.DataFrame(zip(gram3), columns=['three'])
three = []
for i in c['three']:
    three.append(','.join(i))
```

```python
d = pd.DataFrame(zip(gram4), columns=['four'])
four = []
for i in d['four']:
    four.append(','.join(i))
```

```python
e = pd.DataFrame(zip(gram5), columns=['five'])
five = []
for i in e['five']:
    five.append(','.join(i))
```

```python
unigram = pd.DataFrame(one, columns=['unigram'])
bigram = pd.DataFrame(two, columns=['bigram'])
trigram = pd.DataFrame(three, columns=['trigram'])
fourgram = pd.DataFrame(four, columns=['fourgram'])
fivegram = pd.DataFrame(five, columns=['fivegram'])
```

```python
n_gram = pd.concat([unigram,bigram,trigram,fourgram,fivegram], axis=1)
n_gram
```

```python
news = pd.concat([brazilnews,n_gram], axis=1)
news
```
