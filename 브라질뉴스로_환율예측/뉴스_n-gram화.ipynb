```python
import nltk
nltk.download('punkt')
import nltk.data
sent_tokenizer=nltk.data.load('tokenizers/punkt/portuguese.pickle')
# from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
nltk.download('stopwords')
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("portuguese")
```

word_tokenize->stopwords->stemming->n-gram 순

```python
def get_ngrams(text, n ):

#     sentences = sent_tokenizer.tokenize(text)
#     tmp = []
    for i in sentences:
        word_tokens = word_tokenize(i)
        result = []
        for token in word_tokens: 
            if token not in stopwords: 
                result.append(token)

        stem_lis = []
        for word in result:
            stem_lis.append(stemmer.stem(word))

        n_grams = ngrams(stem_lis,n)

        result = [';'.join(grams) for grams in n_grams]
    return ','.join(result)
```

```python
gram1 = []
gram2 = []
gram3 = []
gram4 = []
gram5 = []

for i in brazilnews['total']:
    sentences = sent_tokenizer.tokenize(i)
    gram1.append(get_ngrams(sentences,1))
    gram2.append(get_ngrams(sentences,2))
    gram3.append(get_ngrams(sentences,3))
    gram4.append(get_ngrams(sentences,4))
    gram5.append(get_ngrams(sentences,5))
```

### 데이터프레임화 한 후 브라질 뉴스와 결합

```python
unigram = pd.DataFrame(zip(gram1), columns=['unigram'])
bigram = pd.DataFrame(zip(gram2), columns=['bigram'])
trigram = pd.DataFrame(zip(gram3), columns=['trigram'])
fourgram = pd.DataFrame(zip(gram4), columns=['fourgram'])
fivegram = pd.DataFrame(zip(gram5), columns=['fivegram'])
```

```python
n_gram = pd.concat([unigram,bigram,trigram,fourgram,fivegram], axis=1)
n_gram
```

```python
news = pd.concat([brazilnews,n_gram], axis=1)
news
```
