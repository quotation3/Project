folha de sao pualo 뉴스 전처리

```python
folha = pd.read_csv('C:\\Users\\user\\Desktop\\globo\\folha_2019.csv')
folha
```

```python
# 이상한 형태의 숫자 수정
folha['date'] = folha['date'].replace('1º','1')
```

```python
date = []
for i in folha['date']:
    date.append(i.replace('1º','1').strip())
```

```python
dates = pd.DataFrame(date)
final = pd.concat([folha,dates], axis=1)
final.drop(['date'], axis='columns', inplace=True)
```

```python
final.rename(index = {'0': 'date'}, inplace = True)
final.columns = ['content', 'main_title', 'sub_title', 'date']        
```

```python
folha_2 = final[['content', 'date', 'main_title', 'sub_title']]
folha_2
```

```python
folha_2['date']  = folha_2['date'].str.strip()
```

```python
# 형태 맞추기 위해 day가 한자리 수인 경우 앞에 0 붙이기
ls = []
for i in folha_2['date']:
    if len(i) == 10:
        ls.append('0'+i)
    else:
        ls.append(i)
```

```python
# 포르투갈어 월형식을 영어 월형식으로 바꾸기
lis = []
for i in ls:
    if 'fev' in i:
        lis.append(i.replace('fev','feb'))
    elif 'abr' in i:
        lis.append(i.replace('abr','apr'))
    elif 'mai' in i:
        lis.append(i.replace('mai','may'))
    elif 'mai' in i:
        lis.append(i.replace('fev','feb'))
    elif 'ago' in i:
        lis.append(i.replace('ago','aug'))
    elif 'set' in i:
        lis.append(i.replace('set','sep'))
    elif 'out' in i:
        lis.append(i.replace('out','oct'))
    elif 'dez' in i:
        lis.append(i.replace('dez','dec'))
    else:
        lis.append(i)
```

```python
dates = pd.DataFrame(lis)
final_2 = pd.concat([folha_2,dates], axis=1)
final_2.drop(['date'], axis='columns', inplace=True)
```

```python
final_2.rename(index = {'0': 'date'}, inplace = True)
final_2.columns = ['content', 'main_title', 'sub_title', 'date']  
folha_3 = final_2[['content', 'date', 'main_title', 'sub_title']]
folha_3
```

```python
folha_3['date'] = pd.to_datetime(folha_3['date'], format='%d.%b.%Y')
```

두 개 언론사 뉴스 합치기

```python
brazilnews = pd.concat([globo,folha_3])
brazilnews = brazilnews[['date','content','main_title','sub_title']]
brazilnews
```

```python
brazilnews.sort_values(by=['date'], axis=0, inplace=True)
```
