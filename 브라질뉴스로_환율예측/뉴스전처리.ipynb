```python
# 형태 맞추기 위해 day가 한자리 수인 경우 앞에 0 붙이기
ls = []
for i in folha_2['date']:
    if len(i) == 10:
        ls.append('0'+i)
    else:
        ls.append(i)
```

```python
# 포르투갈어 월형식을 영어 월형식으로 바꾸기
lis = []
for i in ls:
    if 'fev' in i:
        lis.append(i.replace('fev','feb'))
    elif 'abr' in i:
        lis.append(i.replace('abr','apr'))
    elif 'mai' in i:
        lis.append(i.replace('mai','may'))
    elif 'mai' in i:
        lis.append(i.replace('fev','feb'))
    elif 'ago' in i:
        lis.append(i.replace('ago','aug'))
    elif 'set' in i:
        lis.append(i.replace('set','sep'))
    elif 'out' in i:
        lis.append(i.replace('out','oct'))
    elif 'dez' in i:
        lis.append(i.replace('dez','dec'))
    else:
        lis.append(i)
```

```python
dates = pd.DataFrame(lis)
final_2 = pd.concat([folha_2,dates], axis=1)
final_2.drop(['date'], axis='columns', inplace=True)
```

```python
final_2.rename(index = {'0': 'date'}, inplace = True)
final_2.columns = ['content', 'main_title', 'sub_title', 'date']  
folha_3 = final_2[['content', 'date', 'main_title', 'sub_title']]
folha_3
```

2018, 2019 합치기

```python
news_2018 = pd.read_csv('C:\\Users\\user\\Desktop\\globo\\brazilnews_2018.csv')
news_2019 = pd.read_csv('C:\\Users\\user\\Desktop\\globo\\brazilnews.csv')
```

```python
len(news_2019)
```

```python
news = pd.concat([news_2018,news_2019],ignore_index=True)
news
```

```python
titles = []
for i in news['main_title']:
    titles.append(i.replace('\n',''))
```

```python
sub = []
for i in news['sub_title']:
    if type(i) != str:
        sub.append(i)
    else:
        sub.append(i.replace('\n',''))
```

```python
titles = pd.DataFrame(titles)
titles.rename(index = {'0': 'main_title_2'}, inplace = True)
```

```python
subtitles = pd.DataFrame(sub)
subtitles.rename(index = {'0': 'sub_title_2'}, inplace = True)
```

```python
df = pd.concat([titles,subtitles], axis=1)
df
```

```python
news_final = pd.concat([news,df], axis=1)
news_final.drop(['main_title'], axis='columns', inplace=True)
news_final.drop(['sub_title'], axis='columns', inplace=True)
```

```python
news_final.columns = ['date', 'content', 'main_title', 'sub_title']
news_final
```

```python
news_final.to_csv('C:\\Users\\user\\Desktop\\globo\\brazilnews_final.csv', encoding='utf-8', index=False)
```
