```python
import pandas as pd
import numpy as np
```

```python
stock = pd.read_csv('C:\\Users\\Inyong Kim\\Desktop\\samsung_stock_price.csv',encoding = 'utf-8')
```

```python
stock.rename(columns = {'Date': 'date'}, inplace = True)
```

```python
from datetime import datetime, timedelta

start_date = datetime.strptime('20010101', '%Y%m%d')
end_date = datetime.strptime('20201005', '%Y%m%d')

str_date_list = []
while start_date.strftime('%Y%m%d') != end_date.strftime('%Y%m%d'): # start_date가 end_date랑 같아지기 전까지 1씩 늘어나면서 리스트에 추가
    str_date_list.append(start_date.strftime('%Y.%m.%d'))
    start_date += timedelta(days=1)
str_date_list.append(end_date.strftime('%Y.%m.%d')) # end_date 포함안됐으므로 따로 추가

date_list=np.array([str_date_list])
all_date  = pd.DataFrame(date_list.T, columns=['date'])

all_date['date'] = pd.to_datetime(all_date['date'], format="%Y.%m.%d")
stock['date'] = pd.to_datetime(stock['date'], format="%Y.%m.%d")

df = pd.merge(all_date,stock, on='date', how='outer') # 기존주가데이터와 전체날짜를 date기준으로 outer 조인
df
```

```python
new_df = df[['date','Close','Volume']]
new_df
```

```python
new_df = new_df.set_index('date')
new_df['Close'] = new_df['Close'].interpolate(method='time')
new_df['Volume'] = new_df['Volume'].interpolate(method='time')
new_df
```

```python
new_df.reset_index(drop=False, inplace=True)
new_df
```

```python
new_df['after_week']=new_df['Close']
new_df
```

```python
new_df.after_week = new_df.after_week.shift(-7)
new_df
```

```python
new_df['label']=''
```

```python
for i in range(len(new_df)):
    if (new_df['Close'][i] - new_df['after_week'][i])/new_df['Close'][i] > 0.02:
        new_df['label'][i] = new_df['label'][i].replace('','1')
    elif (new_df['Close'][i] - new_df['after_week'][i])/new_df['Close'][i] < -0.02:
        new_df['label'][i] = new_df['label'][i].replace('','0')
```

```python
up = new_df['label'] == '1'
up_table = new_df[up]
up_table
```

```python
down = new_df['label'] == '0'
down_table = new_df[down]
down_table
```
