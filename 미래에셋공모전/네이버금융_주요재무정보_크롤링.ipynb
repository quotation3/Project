```python
def naver_fin(stock_n):
    url = 'https://finance.naver.com/item/main.nhn?code={}'.format(stock_n)
    tables = pd.read_html(url, encoding='euc-kr',index_col=None, header=0)
    df = tables[3]
    new_df=df.loc[[2,3,4,8],['주요재무정보','최근 분기 실적','최근 분기 실적.1','최근 분기 실적.2','최근 분기 실적.3','최근 분기 실적.4']]
    
    asset_1.append(float(new_df.iloc[0][1]))
    profit_1.append(float(new_df.iloc[1][1]))
    net_profit_1.append(float(new_df.iloc[2][1]))
    debt_1.append(float(new_df.iloc[3][1]))

    asset_2.append(float(new_df.iloc[0][2]))
    profit_2.append(float(new_df.iloc[1][2]))
    net_profit_2.append(float(new_df.iloc[2][2]))
    debt_2.append(float(new_df.iloc[3][2]))

    asset_3.append(float(new_df.iloc[0][3]))
    profit_3.append(float(new_df.iloc[1][3]))
    net_profit_3.append(float(new_df.iloc[2][3]))
    debt_3.append(float(new_df.iloc[3][3]))

    asset_4.append(float(new_df.iloc[0][4]))
    profit_4.append(float(new_df.iloc[1][4]))
    net_profit_4.append(float(new_df.iloc[2][4]))
    debt_4.append(float(new_df.iloc[3][4]))
    
    asset_5.append(float(new_df.iloc[0][5]))
    profit_5.append(float(new_df.iloc[1][5]))
    net_profit_5.append(float(new_df.iloc[2][5]))
    debt_5.append(float(new_df.iloc[3][5]))
```
